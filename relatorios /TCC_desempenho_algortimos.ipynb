&emsp;Os testes foram realizados com 2 conjuntos de streams geradas através do MOA, ambos de mudança abrupta e gradual, o primeiro par de streams, utilizado no tuning do PSI foi criado utilizando uma seed rândomica, enquanto o segundo, utilizado na comparação dos detectores, foi criado com outra.

###Primeiro arquivo, testes de tuning para os parametros de threshold de detecção de aviso e de mudança

```python
df = pd.read_csv(path)

#Tratativa para incluir índices representativos do que cada linha representa, sendo eles os parâmetros utilizados nos testes
contador = 0
parametros = ['(2,5)',
'(2,10)',
'(2, 15)',
'(2, 20)',
'(2, 30)',
'(5, 10)',
'(5, 15)',
'(5, 20)',
'(5,30)',
'(10, 15)',
'(10, 20)',
'(10,30)',
'(15, 20)',
'(15, 30)',
'(20, 30)']

novo_indice = []
tipo = ''
for i in df.index:
    if i % 2 == 0:
      tipo = 'Abrupt'
      if i != 0:
        contador+=1
    else:
      tipo = 'Gradual'
    param = tipo + parametros[contador]
    novo_indice.append(param)

df.index = novo_indice
#Convertendo os dados do dataframe para poder realizar eventuais cálculos e comparações dos valores
df = df.astype(float)
df
```

&emsp;Separando os dados úteis para a avaliação em um novo dataframe, os dados escolhidos são:
- A quantidade de mudanças detectadas pelo algoritmo
- A quantidade de avisos detectados pelo algoritmo
- A quantidade de mudanças verdadeiras na stream
- O tempo de atraso para detecção
\
&emsp;Também foi adicionado um campo que representa qual a porcentagem das mudanças detectadas eram de fato mudanças verdadeiras.

```python
#Separando as colunas relevantes no novo dataframe
new_df = df[['Detected changes', 'Detected Warnings', 'True changes', 'Delay detection', 'True changes detected']]

#Adicionando novo campo representando quantos dos acertos detectados eram mudanças de conceito reais.
accuracy = (new_df['True changes']/new_df['Detected changes']) * 100
new_df.loc[:,'True Change Accuracy'] = accuracy
new_df
```

&emsp;Com base nos resultados acima, foi escolhido aplicar um filtro, mantendo assim somente os casos onde a acurácia foi superior a média da coluna 'True Change Accuracy',

```python
media = np.mean(new_df['True Change Accuracy'])
media
```

```python
min_acc = (new_df['True Change Accuracy'] >= media)
min_acc_df = new_df[min_acc]
min_acc_df = min_acc_df.sort_values(by='True Change Accuracy', ascending=False)
min_acc_df
```

Ordenando os resultados pela taxa de precisão de mudanças verdadeiras, é possível perceber que as primeiras posições estão ocupadas pelos testes com parametro de valor de detecção de mudança igual a 30, visualizando através dos 5 primeiros resultados conforme a tabela seguinte

```python
min_acc_df.head(5)
```

```python
acc = min_acc_df.loc['Gradual(2, 30)', 'True Change Accuracy']
print(f"True Change Accuracy para Gradual(2, 30): {acc}")
```

Os dois 'piores' resultados para a taxa de acurácia foram obtidos pelos testes com parametros (2, 30) e (15, 30) ambos com streams graduais. Mesmo assim se mantiveram acima da taxa de precisão de 3.83%, valor acima da média das amostras.

Sendo assim o conjunto de pares contendo o valor 30 para a mudança detectada foi escolhido para refinar a parametrização do tamanho da janela, visto que com tamanho 100 não foi possível detectar situções de aviso, e nem identificar corretamente a ocorrência de mudanças de conceito.

###Segundo arquivo, testes de tuning para o arquivo com os testes de tamanho da janela

&emsp;Com os resultados obtidos no teste anterior, foram definidos parâmetros fixos correspondentes aos melhores resultados para determinar o melhor valor possivel para o tamanho da janela.

```python
df_window = pd.read_csv(path_window)
contador = 0

#Assim como no teste anterior, aplica-se uma tratativa para nomear os índices conforme os parametros testados
parametros = [
'(100, 2, 30)',
'(100, 5,30)',
'(100, 10,30)',
'(100, 15, 30)',
'(100, 20, 30)',
'(500, 2, 30)',
'(500, 5,30)',
'(500, 10,30)',
'(500, 15, 30)',
'(500, 20, 30)',
'(1000, 2, 30)',
'(1000, 5,30)',
'(1000, 10,30)',
'(1000, 15, 30)',
'(1000, 20, 30)'
]

novo_indice = []
tipo = ''
for i in df_window.index:
    if i % 2 == 0:
      tipo = 'Abrupt'
      if i != 0:
        contador+=1
    else:
      tipo = 'Gradual'
    param = tipo + parametros[contador]
    novo_indice.append(param)

df_window.index = novo_indice
df_window = df_window.astype(float)
df_window
```

&emsp;Separando do novo dataframe somente os dados relevantes como mencionado anteriormente na tratativa do primeiro teste, obtem-se os dados do teste do tamanho da janela

```python
new_df_window = df_window[['Detected changes', 'Detected Warnings', 'True changes', 'Delay detection', 'True changes detected']]
accuracy = (new_df_window['True changes']/new_df_window['Detected changes']) * 100
new_df_window.loc[:,'True Change Accuracy'] = accuracy
new_df_window
```

&emsp;Para os casos de janelas com tamanho 100, novamente os resultados não foram bons e além de não gerarem avisos, detectaram um valor muito acima do real em relação a quantidade de mudanças reais na stream.

&emsp;Porém para os casos de janela igual a 500, apesar da melhora em relação a detecção de falsos positivos, obtendo uma quantidade mais próxima da real, o atraso para a detecção aumentou significativamente.

```python
tunning_final = (new_df_window['True Change Accuracy'] >= 50)
tunning_final_df = new_df_window[tunning_final]
tunning_final_df.head(10)
```

&emsp;Observa-se que nos testes onde o tamanho da janela é igual a 1000, obtêm-se resultados melhores. Os testes apresentaram detecção de mudança de conceito constante que não estava presente nos testes com janelas diferentes, apresentando um resultado condizente com o valor utilizado como parametro para detecção de mudanças.

&emsp;Além disso, os tempos de atraso para detecção não mudaram entre os diferentes valores de threshold para aviso e detecção real de mudança, mas sim entre os tipos de stream, apresentando tempos constantes de 96.3 para a stream gradual e um tempo de 85.7	para a abrupta.

&emsp;Foi escolhida como parametrização padrão final, o tamanho da janela sendo igual a 1000, o limite de aviso igual a 20 e o alerta de  detecção igual a 30. Com o valor do aviso podendo alterar conforme contexto onde for utilizado e sem prejudicar o resultado.

###Testes comparativos com os outros detectores

&emsp;Conforme comentado no teste anterior, foi escolhido o melhor parametro para realizar testes com o PSI para comparar com os principais detectores de mudança de conceito, sendo esses DDM, EDDM, ADWIN, CUSUM e Page-Hinkley.

```python
#Aqui fazemos a tratativa dos índices para facilitar a leitura e identificação dos dados nos dataframes

#Principais algoritmos com parametros default
df_others = pd.read_csv(path_others)

contador = 0
parametros = [
'(DDM)',
'(EDDM)',
'(ADWIN)',
'(CUSUM)',
'(PAGE-HINKLEY)'
]

novo_indice = []
tipo = ''
for i in df_others.index:
    if i % 2 == 0:
      tipo = 'Abrupt'
      if i != 0:
        contador+=1
    else:
      tipo = 'Gradual'
    param = tipo + parametros[contador]
    novo_indice.append(param)

df_others.index = novo_indice
df_others = df_others.astype(float)

#PSI com os parametros selecionados
df_tuned = pd.read_csv(path_tuned)

contador = 0
parametros = [
'(PSI 1000, 20, 30)',
]

novo_indice = []
tipo = ''
for i in df_tuned.index:
    if i % 2 == 0:
      tipo = 'Abrupt'
      if i != 0:
        contador+=1
    else:
      tipo = 'Gradual'
    param = tipo + parametros[contador]
    novo_indice.append(param)

df_tuned.index = novo_indice
df_tuned = df_tuned.astype(float)

# junção dos dataframes
df_detectors = pd.concat([df_others, df_tuned])
df_detectors
```

&emsp;A seguir, obtém-se o novo dataframe de comparação dos detectores, filtrando os campos desejados conforme feito nos testes anteriores.

```python
new_df_detectors = df_detectors[['Detected changes', 'Detected Warnings', 'True changes', 'Delay detection', 'True changes detected']]
# accuracy = (new_df_detectors['True changes']/new_df_detectors['Detected changes']) * 100


new_df_detectors.loc[:,'True Change Accuracy'] = accuracy
new_df_detectors
```

Referente à coluna 'True Change Accuracy', como a mesma representa a quantidade, em porcentagem, de quantas mudanças detectadas de fato são mudanças verdadeiras na stream, os dois casos do PAGE-HINKLEY apresentaram valores superiores a 100% pois, na verdade, deixaram de detectar mudanças de conceito.

###Conclusão

&emsp;Conforme o dataframe acima, é possível perceber que aumentando o valor da janela do PSI para além do que foi inicialmente estipulado como padrão, janela de tamanho 100, o método é capaz de detectar com maior precisão a quantidade de mudanças ocorridas. \
&emsp;Podemos notar que o PSI obteve melhores resultados que os seguintes detectores no quesito de acurácia: EDDM, ADWIN e PAGE-HINKLEY, esse inclusive deixou de detectar mudanças que ocorreram na stream. Quanto ao atraso para a detecção, ficou somente atrás do ADWIN e CUSUM, respectivamente.\
